### Importing Libraries

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
r = requests.get('https://www.gkseries.com/general-knowledge/geography/geo-tectonics/geography-mcqs')
url = 'https://www.gkseries.com/general-knowledge/geography/geo-tectonics/'
```

```python
soup = BeautifulSoup(r.content,'html.parser')
```

### Extracting Links

```python
links = []
link_component = soup.find_all('li',{'itemprop':'position'})
for link in link_component:
    links.append(url+link.find('a')['href'])
links = links[0:-1]
```

### Record = {Question, Option1, Option2, Option3, Option4, Answer}

```python
record = []
for link in links:
    
    r1 = requests.get(link)
    soup2 = BeautifulSoup(r1.content,'html.parser')
    question_component = soup2.find_all('div',{'class':'mcq'})
    
    for component in question_component :
    
        #Question Number
        q_no = component.find('span')
        q_no = q_no.text
    
        #Question
        q = component.find('div',{'class':'question-content'})
        question = q.text.strip()[7:].strip()
    
        #Options
        Options = component.find_all('div',{'itemprop':'text'})
        Option1 = Options[0].text[1:].strip()
        Option2 = Options[1].text[1:].strip()
        Option3 = Options[2].text[1:].strip()
        Option4 = Options[3].text[1:].strip()
    
        #Answer
        ans = component.find('blockquote')
        answer= ans.text.strip().split(' ')[2]
    
        record.append((q_no,question,Option1,Option2,Option3,Option4,answer))
```

```python
len(record)
```

```python
data = pd.DataFrame(record,columns=['Question NUmber','Question','Option A','Option B','Option C','Option D','Answer'])
data.reset_index(drop=True,inplace = True)
```
